```python
import pandas as pd
import seaborn as sns

carros = sns.load_dataset('mpg')
```

```python
# Mostra as primeiras 5 linhas do conjunto de dados
carros.head()

# Mostra informações básicas sobre as colunas do conjunto de dados
carros.info()

# Mostra a estatística descritiva do conjunto de dados
carros.describe()
```

```python
# Cria um histograma da distribuição de consumo de combustível dos carros
sns.histplot(data=carros, x='mpg')

# Cria um scatterplot entre o peso e o consumo de combustível dos carros
sns.scatterplot(data=carros, x='weight', y='mpg', hue='origin')
```

```python
# Salva o conjunto de dados em um arquivo CSV
carros.to_csv('carros.csv', index=False)
```
